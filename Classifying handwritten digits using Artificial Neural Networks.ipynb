# Classifying handwritten digits using artificial neural networks

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
# https://keras.io/api/datasets/

(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[0]
```

```python
y_train
```

```python
plt.imshow(X_train[0])
print(y_train[0])
plt.axis('off');
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
X_train.shape
```

```python
X_train_flattened = X_train.reshape(60000, 784)
X_train_flattened.shape
```

```python
X_train_flattened[0]
```

```python
X_test_flattened = X_test.reshape(10000, 784)
X_test_flattened.shape
```

```python
# We need this for the next cell
X_train_flattened[0].shape
```

```python
# ANN with only input and output layers

ann1 = keras.Sequential([    
                         keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
                         ])

ann1.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

ann1.fit(X_train_flattened, y_train, epochs=5)
```

```python
ann1.evaluate(X_test_flattened, y_test)
```

```python
plt.imshow(X_test[0])
print(y_test[0])
plt.axis('off');
```

```python
y_pred = ann1.predict(X_test_flattened)
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
y_pred_labels = [np.argmax(i) for i in y_pred]
```

```python
from sklearn.metrics import confusion_matrix 

cm = confusion_matrix(y_test, y_pred_labels)
cm
```

```python
import seaborn as sn

sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual');
```

```python
# Use a flatten layer

ann2 = keras.Sequential([
                         keras.layers.Flatten(input_shape=(28, 28)),
                         keras.layers.Dense(10, activation='sigmoid')
                         ])

ann2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

ann2.fit(X_train, y_train, epochs=5)
```

```python
ann2.evaluate(X_test, y_test)
```

```python
# Use a hidden layer

ann3 = keras.Sequential([
                         keras.layers.Dense(20, input_shape=(784,), activation='relu'),
                         keras.layers.Dense(10, activation='sigmoid')
                         ])

ann3.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

ann3.fit(X_train_flattened, y_train, epochs=5)
```

```python
ann3.evaluate(X_test_flattened, y_test)
```
